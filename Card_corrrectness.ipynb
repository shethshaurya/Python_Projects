```
num = int(input("16 digit card number: "))
num=str(num)
```

```
def Card_authenticity(num):

  x=[]
  for i in range(len(num)):
    if i%2==0:
      x.append(num[i])

  for i in range(len(x)):
    x[i]=(str(int(x[i])*2))

  for i in range(len(x)):
    if len(x[i])==1:
      x[i]=int(x[i])
    elif len(x[i])==2:
      x[i]=int(int(x[i][0])+int(x[i][1]))

  sum=0
  for i in range(len(num)):
    if i%2!=0:
      sum+=int(num[i])

  for i in range(len(x)):
    sum+=x[i]

  if sum%10==0:
    print('The number entered is correct')
  else:
    print('The number entered is incorrect')
```

```
Card_authenticity(num)
```
